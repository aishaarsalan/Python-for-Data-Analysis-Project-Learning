Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
print(schools.info())
```

## BEST SAT MATHS RESULT IN NYC SCHOOLS

```python
def minimum_score(percentage, max_score):
    return (percentage / 100) * max_score
max_possible_score = 800
best_math_threshold = minimum_score(80, max_possible_score)
print(f"The best math results must be at least: {best_math_threshold}")
```

```python
best_math_schools = schools[schools['average_math'] >= 640.0][['school_name', 'average_math']]
best_math_schools = best_math_schools.sort_values(by='average_math', ascending=False)
print(best_math_schools)
```

The best math results are at least 80% of the maximum possible score of 800, the calculation shows that to become the best math, the result must be at least scored 640.0. 

Out of a total of 375 schools, 10 schools have demonstrated strong performance in their math programs with scores above the minimum threshold 640. Stuyvesant High School has the highest average math score of 754, with a margin of 40 points over the second-ranked Bronx High School of Science. The lowest score close to the minimum threshold is from Eleanor Roosevelt High School which achieved a score of 641.

## TOP 10 Performing Schools

```python
schools["total_SAT"] = schools[["average_math", "average_writing", "average_reading"]].sum(axis=1) 
top_10_schools = schools[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False).head(10)
print(top_10_schools)
```

The data presented highlights the top 10 performing high schools based on their combined SAT scores. This ranking provides insight into the academic excellence and competitiveness of these institutions and demonstrating their ability to prepare students for standardized testing and higher education. 

At the top of the list is Stuyvesant High School which is achieving an impressive combined SAT score of 2144, making it the highest performer. Following by Bronx High School of Science and Staten Island Technical High School with 2041 points. The High School of American Studies at Lehman College ranks fourth with 2013 points. Eleanor Roosevelt High School and High School for Mathematics, Science, and Engineering are in the 9th and 10th with 1889 points. The score range among the top 10 is relatively narrow, spanning from 2144 to 1889, indicating a highly competitive academic landscape. 

In summary, these leading schools demonstrate a deep dedication to academic success, offering students the tools and programs needed to perform well on standardized tests. With specialized courses, selective admissions, and challenging classes, they consistently achieve high SAT scores, solidifying their status as some of the top educational institutions in the nation.

## Identifying the Borough with the Largest Standard Deviation in Combined SAT Score Changes

```python
borough_stats = schools.groupby("borough").agg(
    num_schools=("school_name", "size"),
    average_SAT=("total_SAT", "mean"),
    std_SAT=("total_SAT", "std")
).reset_index()
largest_std_dev = borough_stats.loc[borough_stats["std_SAT"].idxmax()]
largest_std_dev = pd.DataFrame([largest_std_dev]).round(2)
print(largest_std_dev)
```

The data indicates that Manhattan has the largest standard deviation of 230.29 in combined SAT score changes among all boroughs. A high standard deviation suggests significant variability in SAT scores across the 89 schools in the borough. This means that while some schools perform exceptionally well, others may have much lower scores and it's leading to a wide range of performance.
